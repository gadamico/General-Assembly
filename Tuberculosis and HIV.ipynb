```python
def m_tuberpred(X, y):

    results = sm.GLM(y, sm.add_constant(X),
                sm.families.Gamma(sm.families.links.log)).fit()
    
    X_test = pd.DataFrame([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
                      columns = ['future'])
    preds = []
    for i in range(10):
        preds.append(y[34] * np.exp(results.params[1]) ** (i + 1))
    print(results.params[1])
    print(pd.concat([X_test, pd.DataFrame(preds)], axis = 1))
```

```python
def t_tuberpred(X, y):

    results = sm.GLM(y, sm.add_constant(X),
                sm.families.Gamma(sm.families.links.log)).fit()
    
    X_test = pd.DataFrame([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
                      columns = ['future'])
    preds = []
    for i in range(10):
        preds.append(y[104] * np.exp(results.params[1]) ** (i + 1))
    print(results.params[1])
    return pd.concat([X_test, pd.DataFrame(preds)], axis = 1)
```

## Exploring Florida

```python
florida = pd.read_csv('IHME_USA_COUNTY_INFECT_DIS_MORT_1980_2014_STATES_A_TO_F_CSV/IHME_USA_COUNTY_INFECT_DIS_MORT_1980_2014_FLORIDA_Y2018M03D27.CSV')

florida = florida.loc[(florida['cause_name'] == 'HIV/AIDS') & (florida['location_name'] != 'Florida'), :]

florida = florida.drop(columns = ['measure_id', 'measure_name', 'location_id', 'FIPS',
                                'cause_id', 'cause_name', 'sex_id', 'age_id', 'age_name',
                                'metric'])

florida['location_name'].value_counts()
```

### Climate Data

Florida climate data from https://en.wikipedia.org/wiki/Climate_of_Florida#/media/File:Florida_K%C3%B6ppen.svg.

```python
fl_clim_dict = {'Miami-Dade County': 'Monsoon', 'Broward County': 'Rainforest/Monsoon/Savanna',
                'Monroe County': 'Savanna',
                'Collier County': 'Monsoon/Savanna/Humid subtropical'}
```

```python
florida['Climate'] = florida['location_name'].map(fl_clim_dict)

florida['Climate'] = florida['Climate'].fillna('Humid subtropical')

florida['Climate'].value_counts()
```

### Population Data

### Economic Data

```python
url4 = 'https://en.wikipedia.org/wiki/List_of_Florida_locations_by_per_capita_income'
requests.get(url4).status_code
```

```python
soup = BeautifulSoup(requests.get(url4).content, 'lxml')

headers = []
table = soup.findAll('table')[2]

for th in table.findChildren('th'):
    headers.append(th.text)
```

```python
headers
```

```python
headers = [header.replace('\n', ' ') for header in headers]
headers
```

```python
fl_monies = []
table = soup.findAll('table')[2]

for td in table.findChildren('td'):
    fl_monies.append(td.text)
```

```python
fl_moniesdf = pd.DataFrame(np.reshape(fl_monies, (69, 7)), columns = headers)
```

```python
fl_moniesdf.loc[fl_moniesdf['Rank'] == '']
```

```python
zeroes = fl_moniesdf.loc[fl_moniesdf['Rank'] == '']

fl_moniesdf = fl_moniesdf.drop(zeroes.index)

fl_moniesdf = fl_moniesdf.reset_index()

fl_moniesdf = fl_moniesdf.drop(['index', 'Population'], axis = 1)
```

```python
fl_moniesdf['County'].value_counts()
```

```python
counties = [county.replace(' ', '') for county in fl_moniesdf['County']]
```

```python
fl_moniesdf['County'] = counties
```

```python
fl_counties = [county.replace('Saint', 'St.') for county in fl_counties]
```

```python
florida_pop['location_name'] = fl_counties
```
