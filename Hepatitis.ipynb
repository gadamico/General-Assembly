```python
def t_heppred(X, y):

    results = sm.GLM(y, sm.add_constant(X),
                sm.families.Gamma(sm.families.links.log)).fit()
    
    X_test = pd.DataFrame([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022,
                           2023, 2024],
                      columns = ['future'])
    preds = []
    for i in range(10):
        preds.append(y[629] * np.exp(results.params[1]) ** (i + 1))
    print(results.params[1])
    return pd.concat([X_test, pd.DataFrame(preds)], axis = 1)
```

```python
soup = BeautifulSoup(requests.get(url2).content, 'lxml')

headers = []
table = soup.findAll('table')[1]

for th in table.findChildren('th'):
    headers.append(th.text)
```

```python
headers
```

```python
headers = [header.replace('\n', ' ') for header in headers]
headers
```

```python
# For the following trick I relied on
# https://www.quora.com/How-can-I-delete-non-alphabetic-characters-in-a-Python-list

import re
headers = [header.replace('note', '') for header in headers]
headers = [re.sub(r'[^a-zA-Z ]', "", header) for header in headers]
headers
```

```python
ca_monies = []
table = soup.findAll('table')[1]

for td in table.findChildren('td'):
    ca_monies.append(td.text)
```

```python
ca_moniesdf = pd.DataFrame(np.reshape(ca_monies, (58, 6)), columns = headers)
ca_moniesdf
```

```python
ca_moniesdf = ca_moniesdf.drop(columns = ['Population'])
```
