```python
def t_resppred(X, y):

    results = sm.GLM(y, sm.add_constant(X),
                sm.families.Gamma(sm.families.links.log)).fit()
    
    X_test = pd.DataFrame([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022,
                           2023, 2024],
                      columns = ['future'])
    preds = []
    for i in range(10):
        preds.append(y[419] * np.exp(results.params[1]) ** (i + 1))
    print(results.params[1])
    return pd.concat([X_test, pd.DataFrame(preds)], axis = 1)
```

### Economic Data

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_Mississippi_locations_by_per_capita_income'
requests.get(url2).status_code
```

```python
soup = BeautifulSoup(requests.get(url2).content, 'lxml')

headers = []
table = soup.findAll('table')[1]

for th in table.findChildren('th'):
    headers.append(th.text)
```

```python
headers
```

```python
headers = [header.replace('\n', ' ') for header in headers]
```

```python
ms_monies = []
table = soup.findAll('table')[1]

for td in table.findChildren('td'):
    ms_monies.append(td.text)
```

```python
ms_moniesdf = pd.DataFrame(np.reshape(ms_monies, (84, 6)), columns = headers)
```

```python
ms_moniesdf
```

```python
zeroes = ms_moniesdf.loc[ms_moniesdf['Rank'] == '']

ms_moniesdf = ms_moniesdf.drop(zeroes.index)

ms_moniesdf = ms_moniesdf.reset_index()

ms_moniesdf = ms_moniesdf.drop(['index', 'Population'], axis = 1)
```

```python
mississippi_pop.head()
```

```python
ms_counties = [county.replace('County', '') for county in mississippi_pop['County']]

mississippi_pop['County'] = ms_counties
```

```python
counties = [county.replace(' ', '') for county in ms_moniesdf['County']]

ms_moniesdf['County'] = counties
```

```python
mississippi_monies = pd.merge(mississippi_pop, ms_moniesdf, how = 'left', on = 'County')
```

```python
mississippi_monies.head()
```

```python
county_dums = pd.get_dummies(mississippi_monies['County'])
mississippi_monies = pd.concat([mississippi_monies, county_dums], axis = 1)
```

```python
climate_dums = pd.get_dummies(mississippi_monies['Climate'])
mississippi_monies = pd.concat([mississippi_monies, climate_dums], axis = 1)
```

```python
mississippi_monies.isnull().sum().sum()
```
